```python
import numpy as np
import os.path as osp

from matplotlib import pyplot as plt
from torchvision.datasets import CIFAR10
```

```python
root = osp.join("..", "data", "CIFAR10")
idx1 = 0
idx2 = 100
lambda_ = 0.6
```

```python
dataset = CIFAR10(root, transform=np.array)
img1, label_idx1 = dataset[idx1]
img2, label_idx2 = dataset[idx2]

# Convert to onehot
label1 = np.zeros((10,))
label1[label_idx1] = 1.0

label2 = np.zeros((10,))
label2[label_idx2] = 1.0

# Mix images
img_mixed = img1 * lambda_ + img2 * (1.0 - lambda_)
label_mixed = label1 * lambda_ + label2 * (1.0 - lambda_)

names = ["Img 1", "Img 2", "Img mixed"]
imgs = [img1, img2, img_mixed]
for name, img in zip(names, imgs):
    plt.figure()
    plt.title(name)
    plt.imshow(np.array(img, dtype=int))

labels = [label1, label2, label_mixed]
for label in labels:
    plt.figure()
    plt.bar(list(range(len(label))), label)
    axes = plt.gca()
    axes.set_ylim([0.0, 1.1])

plt.show()
```
